### Scale Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train = train.loc[:, train.columns != "target"]
y_train = train.loc[:, train.columns == "target"]
```

```python
X_test = test.loc[:, test.columns != "target"]
y_test = test.loc[:, test.columns == "target"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Since there were only 300 training observations, which would not give me usable results if the training set was split into a validation set. Five fold cross validation on the training set was used to see how the model performed. The mean accuracy and the standard deviation of the 5 folds was taken and compared between models. 

### Linear SVM with SVC OneVsRestClassifier

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
lsvc = OneVsRestClassifier(SVC(kernel="linear"))
```

```python
parameters = {'estimator__C':[1,10,100,1000]}
```

```python
grid = GridSearchCV(lsvc, parameters, cv = 5)
```

```python
grid.fit(X_train_scaled, y_train.values.ravel())
```

```python
grid.best_params_, grid.best_score_
```

```python
model1 = OneVsRestClassifier(SVC(kernel="linear", C = 1000))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def met_met(x):
    print(x)
    print("Mean Accuracy:", x.mean())
    print("Standard Deviation of Accuracy:", x.std())
```

```python
score1 = cross_val_score(model1, X_train_scaled, y_train,cv = 5)
met_met(score1)
```

Based on the information given from the cross validation conducted, model 2 is the best model. This is an SVM model using SVC and OnevsOneClassifier. Model 2 had a C  of 10 and gamma of 5. Model 2 had the best mean accuracy out of all of the models and the second lowest standard deviation. The low standard deviation shows that the accuracy didnâ€™t move very much when tested on new data the model had not seen before. By having the highest mean accuracy and the second lowest standard deviation model 2 was the clear pick.
